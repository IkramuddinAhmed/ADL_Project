```python
#Removing all rows with disasters not relevant to our project
h = df[df["incident_type"].str.contains("Fire") == False]
h = h[h["incident_type"].str.contains("Other") == False]
h = h[h["incident_type"].str.contains("Toxic Substance") == False]
h = h[h["incident_type"].str.contains("Drought") == False]
h = h[h["incident_type"].str.contains("Dam/Levee Break") == False]
h = h[h["incident_type"].str.contains("Snow") == False]
h = h[h["incident_type"].str.contains("Severe Ice Storm") == False]
h = h[h["incident_type"].str.contains("Freezing") == False]
h = h[h["incident_type"].str.contains("Coastal Storm") == False]
h = h[h["incident_type"].str.contains("Fishing Losses") == False]
h = h[h["incident_type"].str.contains("Mud/Landslide") == False]
h = h[h["incident_type"].str.contains("Human Cause") == False]
h = h[h["incident_type"].str.contains("Terrorist") == False]
h = h[h["incident_type"].str.contains("Chemical") == False]
h = h[h["incident_type"].str.contains("Biological") == False]
h = h[h["incident_type"].str.contains("Typhoon") == False]
h = h[h["incident_type"].str.contains("Severe Storm(s)", regex=False) == False]
```

```python
#Dropping all columns not required for Time series analysis
i = h.drop(['fema_declaration_string', 'disaster_number', 
       'declaration_type', 'declaration_date',
       'declaration_title', 'ih_program_declared', 'ia_program_declared',
       'pa_program_declared', 'hm_program_declared',
       'incident_end_date', 'disaster_closeout_date', 'fips', 'place_code',
       'designated_area', 'declaration_request_number', 'hash', 'last_refresh',
       'id'], axis = 1)
```

```python
j = i.drop(['index'], axis=1)
```

```python
j.tail()
```

```python
j = j.rename(columns={"fy_declared":"Year"})
j = j.rename(columns={"incident_begin_date":"Date"})
j = j.rename(columns={"incident_type":"Calamity"})
j = j.rename(columns={"state":"State"})
```

```python
j.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
j['State_Label'] = encoder.fit_transform(j['State'])
j.head(20)
```

```python
new_cols = ["Date","Year","State","State_Label","Calamity"]
o=j.reindex(columns=new_cols)
o.head(10)
```
